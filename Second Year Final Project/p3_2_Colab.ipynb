```python
import pandas as pd
df = pd.read_csv('inaug_speeches.csv', encoding = 'latin1')
```

```python
from transformers import pipeline
classifier = pipeline('zero-shot-classification', model='roberta-large-mnli')
```

```python
quant_scores = []
optimistic_scores = []

for speech in df['text']:
    # classify the speech using the pipeline
    result = classifier(speech, ['quantitative reasoning skills', 'optimistic'])
    # save the scores in the appropriate lists
    quant_scores.append(result['scores'][0])
    optimistic_scores.append(result['scores'][1])

# add the new columns to the dataframe
df['QuantLLMQuant'] = quant_scores
df['Optimistic'] = optimistic_scores
```

```python
df.to_csv('inaug_speeches_LLM.csv', index=False)
```
